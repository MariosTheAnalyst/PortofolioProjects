```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\thoma\Downloads\WorldHappiness_Corruption_2015_2020.csv')
df.head()
```

```python
for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print ('{}-{}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
pd.set_option('display.max_rows', None)
```

```python
plt.scatter(x= df['gdp_per_capita'], y=df['happiness_score'])
plt.title('Scatter plot with GDP per capita vs Happiness_score')
plt.xlabel('gdp_per_capita')
plt.ylabel('happiness_score')
plt.show()
```

```python
sns.regplot(x='gdp_per_capita', y='happiness_score', data=df, scatter_kws= ({'color':'red'}), line_kws={'color':'blue'})
```

```python
df.corr('pearson')
```

```python
Correlation_matrix = df.corr(method='pearson')
sns.heatmap(Correlation_matrix, annot= True)
plt.show()
```

```python
correlation_mat=df.corr()
corr_pairs= correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr= sorted_pairs[(sorted_pairs>0.5)]
high_corr
```
